```python
from sympy import symbols, Eq, solve, init_printing, collect, Matrix

init_printing(use_latex=True)

k, k_t, d, d_t, l_t, l_a, m, I, C_t, C_y, U, y_ddot, y_dot, y, theta_ddot, theta_dot, theta = symbols('k, k_{\\theta} d d_{\\theta} l_{\\theta} l_{\\alpha} m I^F C_{\\theta} C_y U \\ddot{y} \\dot{y} y \\ddot{\\theta} \\dot{\\theta} \\theta') 

L = C_t * U ** 2 * theta + C_y * U * y_dot

eq1 = Eq(m * y_ddot - m * l_t * theta_ddot + d * y_dot + k * y, L)
eq2 = Eq(I * theta_ddot - m * l_t * y_ddot + d_t * theta_dot + k_t * theta, l_a * L)

solution = solve([eq1, eq2], [y_ddot, theta_ddot])

collected = collect(solution[y_ddot], [y_dot, theta, y, theta_dot])



M = Matrix([
    [0, 0, 1, 0],
    [0, 0, 0, 1],
    [C_t * U * U * (l_a + l_t) - k_t, -l_t * k, -d_t, C_y * U * (l_a + l_t) - l_t * d],
    [C_t * U * U + m * l_t * (l_a * C_t * U * U - k_t) / I, -k, -m * l_t * d_t / I, C_y * U * (1 + m * l_t * l_a / I) - d],
])

# M.rescale_row(2, I - m * l_t * l_t)
# M.rescale_row(3, m - m * m * l_t * l_t / I)

M.eigenvals()
```
